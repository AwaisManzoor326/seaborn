```python
import seaborn as sns 
from matplotlib import pyplot as plt
```

**Histogram**

Histograms represent the data distribution by forming bins along with the range of the data and then drawing bars to show the number of observations that fall in each bin.In Seaborn we use distplot() function to plot histograms.

```python
sns.set_theme('paper')
```

```python
df = sns.load_dataset('iris')
sns.displot(df['petal_length'], kde = False)
```

**Bar Plot**

Seaborn supports many types of bar plots and you will see a few of them here. Here, as mentioned in the introduction we will use both seaborn and matplotlib together to demonstrate several plots.

```python
sns.set_context('paper')

titanic = sns.load_dataset('titanic')
sns.barplot(x='embark_town', y='age', data=titanic, palette='PuRd', errorbar=None)
plt.legend()
plt.show()
print(titanic.columns)
```

```python
sns.barplot(x='sex', y='survived', hue='class', data=titanic, palette='PuRd', order=['male', 'female'], capsize=0.5, saturation=8, errcolor = 'gray', errwidth = 2, errorbar=None)
plt.legend()
plt.show()
```

To draw a horizontal plot pass ‘h’ to the parameter, orient of the barplot function

```python
sns.barplot(x = 'age', y = 'embark_town', data = titanic,palette = 'PuRd', orient='h')
plt.show()
```

**Count plot**

The count plot can be thought of as a histogram across a categorical variable.

```python
sns.countplot(x='class', hue='who', data=titanic, palette='magma')
plt.title('Survivors')
plt.show()
```

**Point Plot**

Point plot is used to show point estimates and confidence intervals using scatter plot glyphs. A point plot represents an estimate of central tendency for a numeric variable by the position of scatter plot points and provides some indication of the uncertainty around that estimate using error bars.

Point plots can be more useful than bar plots for focusing comparisons between different levels of one or more categorical variables.

```python
tips_data = sns.load_dataset('tips')
sns.pointplot(data=tips_data, x='day', y='tip')
plt.show()
```

```python
print(tips_data.head())
```

```python
sns.pointplot(x='time', y='total_bill', hue='smoker', data=tips_data, palette='Accent')
plt.show()
```

**Joint Plot**

Joint Plot draws a plot of two variables with bivariate and univariate graphs. It uses the Scatter Plot and Histogram. Joint Plot can also display data using Kernel Density Estimate (KDE) and Hexagons. We can also draw a Regression Line in Scatter Plot.

```python
sns.set_style('dark')
sns.jointplot(x='total_bill', y='tip', data=tips_data)
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips_data, kind='reg', color='green')
```

```python
sns.set_context('talk')
sns.jointplot(x='total_bill', y='tip', data=tips_data, kind='kde', color='green')
```

```python
sns.set_context('paper')
sns.jointplot(x='total_bill', y='tip', data=tips_data, kind='hex', color='purple')
```
